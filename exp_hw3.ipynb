```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("homework_3.1.csv")
t0 = 50  # you can manually set this or infer from context
df['event'] = (df['time'] >= t0).astype(int)
df['time_centered'] = df['time'] - t0
df['interaction'] = df['time_centered'] * df['event']

X = sm.add_constant(df[['event', 'time_centered', 'interaction']])
y = df['Y']
model = sm.OLS(y, X).fit()
print(model.summary())
```
